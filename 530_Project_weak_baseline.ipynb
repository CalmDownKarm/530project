```python
from pathlib import Path
dir = Path("./data/")
list(dir.glob("*"))
```

```python
tweets_dir = dir/"tweets"
```

```python
import pandas as pd
```

```python
from tqdm.auto import tqdm
```

```python
all_tweets = []
files = list(tweets_dir.glob("*csv"))
for filename in tqdm(files):
  df = pd.read_csv(filename)
  game_id, team_nick = filename.stem.split("_")
  df["game_id"] = game_id
  df["team_nick"] = team_nick
  all_tweets.append(df)
```

```python
df = pd.concat(all_tweets)
```

```python
df.to_csv(dir/"all_tweets.csv")
```

```python
tweet_counts = df.groupby(["game_id", "team_nick"]).count().reset_index()[["game_id", "team_nick", "id"]]
```

```python
all_games = pd.read_csv(dir/"games.csv")
all_games['GAME_ID'] = all_games["GAME_ID"].astype(str)
```

```python
games_before_2021 = all_games[all_games['GAME_DATE_EST'] < '2021']
```

```python
team_counts = [0,0,0]
two_team_games = []
for ind, game in all_games.iterrows():
  num_tweets = tweet_counts[tweet_counts['game_id'] == game['GAME_ID']]
  num_rows = num_tweets.shape[0]
  team_counts[num_rows] += 1
  if num_rows == 2:
    two_team_games.append(game['GAME_ID'])
for string_, count in zip(['No tweets', "1 team", "2 team"], team_counts):
  print(string_, count)
```

```python
all_games[all_games['GAME_ID'].isin(two_team_games)].to_csv(dir/"games_with2_team_tweets.csv")
```

```python
df[df['game_id'].isin(two_team_games)].to_csv(dir/"Tweets_for_games_with2_team_tweets.csv")
```
